```python
import pandas as pd
```

```python
#========================================
#loading all csv files into dataframes
#========================================
base_path = r'D:\Guvi_DS_BC\Mini_Projects\InternationalDebtAnalysisSystem\data\Raw'
df_raw = pd.read_csv(f"{base_path}\\IDS_ALLCountries_Data.csv",encoding='latin-1')
df_country_meta = pd.read_csv(f"{base_path}\\IDS_CountryMetaData.csv",encoding='latin-1')
df_footnote = pd.read_csv(f"{base_path}\\IDS_FootNoteMetaData.csv",encoding='latin-1')
df_series_meta = pd.read_csv(f"{base_path}\\IDS_SeriesMetaData.csv",encoding='latin-1')
df_country_series = pd.read_csv(f"{base_path}\\Country-Series - Metadata.csv",encoding='latin-1')
```

```python
print("df_raw:", df_raw.shape)
print("df_country_meta:", df_country_meta.shape)
print("df_series_meta:", df_series_meta.shape)
print("df_country_series:", df_country_series.shape)
print("df_footnote:", df_footnote.shape)
```

```python
df_raw.isnull().sum()
```

```python
df_raw.drop(['Counterpart-Area Name','Counterpart-Area Code'],axis=1,inplace=True)
```

```python
df_raw.drop([str(year) for year in range(2025,2033)],axis=1,inplace=True)
```

```python
df_raw.sample(5,random_state=43)
```

```python
df_raw.isnull().sum()
```

```python
df_raw.dropna(subset='Country Name',inplace=True)
```

```python
year_cols = [str(year) for year in range(2000,2025)]
df_raw.dropna(subset=year_cols, how='all',inplace=True)
```

```python
df_raw.shape
```

```python
print(df_raw[['Country Name', 'Country Code', 'Series Name', 'Series Code']].isnull().sum())
```

```python
print("Shape:", df_raw.shape)
print("Unique Countries:", df_raw['Country Name'].nunique())
print("Unique Indicators:", df_raw['Series Name'].nunique())
```

```python
#Converting wide format to long format

df_long = pd.melt(df_raw, id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'], var_name='year', value_name='debt_value')
print(df_long.shape)
```

```python
df_long.head()
```

```python
df_long.dropna(subset='debt_value',inplace=True)
df_long.shape
```

```python
df_long.isnull().sum()
```

```python
df_long.dtypes
```

```python
df_long['year'] = df_long['year'].astype(int)
df_long.dtypes
```

```python
#Saving the cleaned data to csv
df_long.to_csv(r'D:\Guvi_DS_BC\Mini_Projects\InternationalDebtAnalysisSystem\data\Processed\cleaned_debt_data.csv', index=False)
print("Saved successfully!")
```
