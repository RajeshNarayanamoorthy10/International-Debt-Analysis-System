```python
import pandas as pd
import os
```

```python
processed_path = r'D:\Guvi_DS_BC\Mini_Projects\InternationalDebtAnalysisSystem\data\Processed'

df_long = pd.read_csv(os.path.join(processed_path,"cleaned_debt_data.csv"),
                      dtype={
                          'Country Name': 'category',
                          'Country Code': 'category',
                          'Series Name': 'category',
                          'Series Code': 'category'
                      })
df_country = pd.read_csv(os.path.join(processed_path,"cleaned_country_meta.csv"))
df_series = pd.read_csv(os.path.join(processed_path,"cleaned_series_meta.csv"))


print("Shape of long DF:",df_long.shape)
print("Shape of country DF:",df_country.shape)
print("Shape of series DF",df_series.shape)
```

```python
print(df_long['Country Code'].head())
print(df_country['Code'].head())
```

```python
df_country['Code'] = df_country['Code'].str.strip()
df_long['Country Code'] = df_long['Country Code'].str.strip()
```

```python
print(df_long['Country Code'].head())
print(df_country['Code'].head())
```

```python
df_long_filtered = df_long[df_long['Country Code'].isin(df_country['Code'])]
df_long_filtered.shape
```

```python
#Finding the statistical summary
df_long.describe()
```

```python
df_long[df_long['debt_value']<0].shape
```

```python
df_long[df_long['debt_value']<0]['Series Name'].value_counts().head(5)
```

```python
#Country_wise Analysis

df_long.groupby('Country Name')['debt_value'].sum()
```

```python
country_debt = df_long_filtered.groupby('Country Name')['debt_value'].sum().reset_index()
country_debt.columns = ['Country Name','Total debt']
country_debt = country_debt.sort_values('Total debt',ascending=False)
#top_10 Countries
country_debt.head(10)
```

```python
#bottom_10_countries
country_debt.tail(10)
```

```python
import plotly.express as px
fig = px.bar(country_debt.head(10), x='Country Name', y='Total debt',
             title='Top 10 Countries by Total Debt',
             color='Total debt',
             text='Total debt')

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(xaxis_tickangle=-45)

fig.show()
```

```python
#Indicator-wise analysis
```

```python
indicator_debt = df_long_filtered.groupby('Series Name')['debt_value'].sum().reset_index()
indicator_debt.columns = ['Series Name','Total Debt Value']
indicator_debt = indicator_debt.sort_values(['Total Debt Value'],ascending=False)
indicator_debt.head(10)
```

```python
fig = px.bar(indicator_debt.head(10), x='Total Debt Value', y='Series Name',
             title='Top 10 Indicators by Total Debt',
             orientation='h',
             color='Total Debt Value',
             text='Total Debt Value')

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(
    yaxis={'categoryorder': 'total ascending'},
    xaxis_range=[0, indicator_debt['Total Debt Value'].max() * 1.2],
    margin=dict(l=300)
)

fig.show()
```

```python

#Year-wise Analysis
yearwise_debt = df_long_filtered.groupby('year')['debt_value'].sum().reset_index()
yearwise_debt.columns = ['Years','Total Debt']
yearwise_debt = yearwise_debt.sort_values('Total Debt',ascending=False)
yearwise_debt.head(10)
```

```python
yearwise_debt = yearwise_debt.sort_values('Years', ascending=True)

fig = px.line(yearwise_debt, x='Years', y='Total Debt',
              title='Global Debt Trend (2000-2024)',
              markers=True,
              )

fig.update_traces(texttemplate='%{text:.2s}', textposition='top center')
fig.show()
```

```python
#Region-wise analysis
```

```python
df_merged = pd.merge(df_long_filtered,df_country,left_on='Country Code',right_on='Code')
df_merged.shape
```

```python
df_merged.head(5)
```

```python
region_debt = df_merged.groupby('Region')['debt_value'].sum().reset_index()
region_debt.columns = ['Region', 'Total Debt']
region_debt = region_debt.sort_values('Total Debt', ascending=False)
print(region_debt)
```

```python
fig = px.pie(region_debt, values='Total Debt', names='Region',
             title='Debt Distribution by Region')
fig.show()
```

```python
Income_group_debt = df_merged.groupby('Income Group')['debt_value'].sum().reset_index()
Income_group_debt.columns = ['Income Group', 'Total Debt']
Income_group_debt = Income_group_debt.sort_values('Total Debt', ascending=False)
print(Income_group_debt)
```

```python
fig = px.pie(Income_group_debt,values='Total Debt',names='Income Group',
             title='Debt Distribution by Region')

fig.show()
```
